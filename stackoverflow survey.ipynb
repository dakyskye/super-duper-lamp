# Some Numpy/Matplotlib Stuff

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = pd.read_csv("survey_results_public.csv", index_col='ResponseId')
```

```python
comps = df.CompTotal.dropna()
```

```python
comps
```

```python
comps = comps.to_numpy()
```

```python
comps
```

```python
ranges = np.array([0, 30000, 60000, 90000, 120000, 150000, 200000, 300000, 500000, comps.max()])
```

```python
range_labels = ['$0-$30,000', '$30,000-$60,000', '$60,000-$90,000', '$90,000-$120,000', '$120,000-$150,000', '$150,000-$200,000', '$200,000-$300,000', '$300,000-$500,000', '> $500,000']
```

```python
bins = np.digitize(comps, ranges)
counts = np.bincount(bins)
```

```python
plt.figure(figsize=(10, 6))
```

```python
bar_positions = np.arange(len(range_labels))
```

```python
counts = counts[1:-1]
```

```python
plt.bar(bar_positions, counts)
```

```python
plt.title('comps')
plt.xlabel('range')
plt.ylabel('respondent count')
plt.xticks(bar_positions, range_labels, rotation=45)
```

```python
plt.show()
```
